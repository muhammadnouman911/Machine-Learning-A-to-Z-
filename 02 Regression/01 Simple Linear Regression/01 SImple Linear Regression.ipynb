# Simple Linear Regression


![Salary vs Year Example](simple_linear_regression.png)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Salary_Data.csv')
X=dataset.iloc[:,0].values
Y=dataset.iloc[:,1].values
# Reshape X to be a 2D array if it is a single feature
# In scikit-learn, the input data for training and prediction should be 2D, even if it contains only one feature. You need to reshape your X array to make it 2D.
X = X.reshape(-1, 1)

```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=1/3, random_state=0)
```

### We will not do feature scaling here because linear regression will take care of feature scaling

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,Y_train) 

Y_pred=regressor.predict(X_test)
```

# Visualization the Graph for Training Dataset

```python
plt.scatter(X_train,Y_train,color='red')
plt.plot(X_train,regressor.predict(X_train),color='blue')
plt.title("Salary vs Experience(Training Set)")
plt.xlabel("Years of Experience")
plt.ylabel('Salary ')
plt.show()
```

# Visualization the graph for Test Dataset

```python
plt.scatter(X_test,Y_test,color='red')
plt.plot(X_train,regressor.predict(X_train),color='blue')
# we don't need to change x_test on above line
plt.title("Salary vs Experience(Test Set)")
plt.xlabel("Years of Experience")
plt.ylabel('Salary ')
plt.show()
```
